```python
#Columns Like Children, Country, Company have missing values , (means no of rows are less).
# the column reservation_status_date is in object form, but we need to convert it to date time.
#(a date column have to be cnverted to date time column)
```

```python
df['reservation_status_date'] = pd.to_datetime(df['reservation_status_date'], format='%d/%m/%Y', dayfirst=True)
# changing the column datatype from object to datetime
# Here, after changing it to datetime format, we are again saving it to the same column.

# ** Since, in the dataset, the dates we somewhat mixed, so to bring it to a Standard Format.
# So, we are Specifying the date format explicitly using the format parameter and set dayfirst=True
# since the dates are in day/month/year format.

# dayfirst=True: Indicates that the day comes before the month in the date strings. 
# This is important for correct parsing when the date format isn't the default (which is month/day/year).
```

```python
df.info() # Checking that it (the column) has been converted to date time format successfully.
```

```python
df.describe()
# By default, only shows the numerical columns.
```

```python
df.describe(include = 'object')
# in the datatable, some 'object' datatype is there in some columns, which are nothing but, Categorical Columns.
# We can see how many unique values are there in the categorical columns
# So, we will use describe() function, which by default is used for numerical columns. 
# But, if we include 'object' then it will give summary statistics of object column.
```

```python
# We can see the number of unique values in each column , also we can see the total number of values in each column.
```

```python
for col in df.describe(include = 'object').columns:
    print(col)
    print(df[col].unique())
    print('-'*50) # Making it visible , by adding a line in between.
# By running a for loop , we can see what are these values 
# So, running a for loop in the columns, which are of object type.
# So, first we are printing the names of the columns of object type, then the unique values under each columns
```

```python
df.isnull().sum()
# Finding Missing Values
# it returns the name of the columns, and it returns the number of values which are missing, in the column.
# it returns the column values which are missing
```

```python
# the columns which have huge number of values missing, we will remove them. Like: agent, company. Since , we can't handle them.
# Since, in country , out of 100000 rows, only about 488 values are missing, so we can drop them.
# So, using df.drop function to remove the columns.(we can also remove rows from this function)
# Since, we are removing a column , so we'll have to write: axis = 1
# Using inplace =  True , since we are changing the original dataframe in it's place.
```

```python
df.drop(['company', 'agent'] , axis = 1 , inplace = True)  # removing the columns : company, agent
```

```python
df.dropna(inplace = True)  # Using this df.dropna to remove the rows with missing values.
```

```python
df.isnull().sum() # So, removed all missing values from data
```

```python
df.describe() # Gives the Summary Statistics of the Numerical Columns
```

```python
# There are many many outliers in the data above.
# Like in the childern's table, babies table 10 isn't possible.
# But, we are not removing these outliers. Since, we don't have any use of these columns here.
# adr = average daily rate is: maximum 5400, which is very high. So, we'll have to remove this outlier. Since, we'll perform analysis on it.
```

```python
df['adr'].plot(kind = 'box')   # Watching the outlier, how much difference of values it gives.
```

```python
df = df[df['adr']<5000]  # Means we are removing values less than 5000 in adr.
```

```python
df['adr'].plot(kind = 'box') # Now, it's good.
```
